<h3>Joining DataFrames , Renaming Coloumns , Multiple Operations on Columns, Conditional Filter on records , Deleting<br><br>Rows & Columns</h3>

```python
soldprice_by_age = x.groupby(['AGE'])['SOLD PRICE'].mean().reset_index()[0:5]
soldprice_by_age_role = x.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()[0:5]
soldprice_comparison = soldprice_by_age_role.merge (soldprice_by_age,
on = 'AGE',
how = 'outer')

soldprice_comparison.rename( columns = { 'SOLD PRICE_x': 'SOLD_PRICE_AGE_ROLE','SOLD PRICE_y': 'SOLD_PRICE_AGE' }, inplace = True )

soldprice_comparison
```

<h1 style = "color : purple ;">Plots</h1>

```python
sn.barplot(x='AGE',y = 'SOLD PRICE', data = soldprice_by_age);
```

```python
sn.barplot(x= 'AGE', y = 'SOLD_PRICE_AGE_ROLE',hue='PLAYING ROLE',data = soldprice_comparison);
```
